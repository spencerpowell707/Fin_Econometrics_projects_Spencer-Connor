```julia
fobs = 8470
sobs = 8497
```

```julia
StatsPlots.plot!(futures[2])
```

```julia
ΔS = diff(spot[2]);
ΔF = diff(futures[2]);
```

```julia
ΔS
```

```julia
ΔF
```

```julia
X = ones(fobs-1,2)
X[:,2] .= ΔF
y = ones(sobs-1,2)
y[:,2] .= ΔS
```

```julia
coefs = StatsKit.MultivariateStats.llsq(ΔF, ΔS; bias=false)
coefs[2]
```

```julia
M = 25
hedgeratios = zeros(M)
ibeg = 1
iend = 60
ii = 1
step = 30

for i=1:step:(step*M)
    coefs = StatsKit.MultivariateStats.llsq(X[ibeg:iend,:], ΔS[ibeg:iend]; bias=false)
    hedgeratios[ii] = coefs[2]
    ibeg += step
    iend += step
    ii += 1
end
```

```julia
StatsPlots.plot(hedgeratios)
```
